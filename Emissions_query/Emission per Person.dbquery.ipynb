```sql
select county_state_name,
       population,
       CAST(
           REPLACE(
               `GHG emissions mtons CO2e`,
               ',',
               ''
           ) AS DOUBLE
       ) / NULLIF(
           CAST(
               REPLACE(
                   population,
                   ',',
                   ''
               ) AS DOUBLE
           ),
           0
       ) AS Emissions_per_person
from emissions_data_2023
order by emissions_per_person desc
LIMIT 10
```
